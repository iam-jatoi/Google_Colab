```python
# DICTIONARY

my_dict = {
    "name": "John",
    "age": 30,
    "city": "New York"}

# get
print(my_dict.get("name"))

#items
print(my_dict.items())

# values
print(my_dict.values())

# pop
print(my_dict.pop("city"))

```

```python
# Tuples ke Methods =
# Tuple ke sirf 2 methods hen
#1. Index - indes se hamen ye maloom hoga ke value kis position pe hy (starts from 0)
#2. Count - count se hamen ye pata chalega ke value tuple

# index
my_tuple = ("apple","banana","kiwi","orange")
x = my_tuple.index("banana")
print(x)

# Count
my_tuple = ("apple","banana","kiwi","orange","kiwi")
y = my_tuple.count("kiwi")
print(y)

```

```python
# List Methhods

my_list = ["Jabbar","Oshaq","Waqar","Sameen","Kareem"]

# append
my_list.append("Jagu Dada")
print(my_list)

# insert
my_list.insert(1, "Chaudhry")
print(my_list)

# copy
x = my_list.copy()
print(x)

# extend = yaani 2 lists ko aapas me merge karna
new_list = ["Jimmy", "Shafkat"]
my_list.extend(new_list)
print("updated list Now", my_list)






```

# Dictionary ko List me Merge karna

```python
# Dictionay ko list ke andar merge kar sakte hen kyun ye dono mutable hen

my_list1 = []

my_dict1 = {
    "name": "John",
    "age": 30,
    "city": "New York"}

my_dict2 = {
    "name": "Yasir",
    "age": 35,
    "city": "Chicago"}

my_list1.append(my_dict1)
my_list1.append(my_dict2)
print(my_list1)


# Agar aapko already existing list mein dictionaries ko merge (combine) karna ho,
# to aap Python ke update() method ka use kar sakte hain.

my_list2 = [{"name": "Ali", "age": 25}]

new_dict1 = {"Country": "Pakistan"}

my_list2[0].update(new_dict1)
print(my_list2)
```

```python
import time
from enum import Enum

class TrafficLight(Enum):
    Red = "Please Stop"
    Yellow = "Get Ready"
    Green = "Go Now"

def traffic_simulator():
    while True:
        for signal in TrafficLight:
            print(f"üö¶ {signal.name} - {signal.value}")
            time.sleep(2)
            print("\n ‚è≥ Waiting For Next Cycle... \n")
            time.sleep(3)


traffic_simulator()
```
